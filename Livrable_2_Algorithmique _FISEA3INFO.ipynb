# II.&nbsp;&nbsp; Modélisation

<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Rappel du problème formel :</h3>


On cherche à déterminer les **tournées optimales de livraison** pour un ensemble de camions partant d’un dépôt, visitant un ensemble de clients, puis revenant au dépôt.  
L’objectif est de **minimiser le coût total** (distance, durée ou énergie), tout en respectant les contraintes opérationnelles de capacités avec un ajout supplémentairede contrainte de fenêtre de temps et de trafic

Nous allons d'abord définir les paramètrese qui regroupent l’ensemble des données connues **avant la résolution** du problème.  
Ils définissent la structure du réseau, les caractéristiques des clients et les propriétés des véhicules:

| Nom | Signification | Exemple |
|---|---|---|
| **V** | Ensemble des clients + dépôt (0 = dépôt) | V = {0,1,2,3} |
| **K** | Ensemble des camions | K = {1,2} |
| **cᵢⱼ** | Coût entre *i* et *j* | c₁₂ = 12|
| **dᵢ** | Demande du client *i* | d₂ = 50|
| **Cmₖ** | Capacité du camion *k* | Cm₁ = 200|
| **[aᵢ, bᵢ]** | Fenêtre temporelle de service | [8h, 12h] |
| **sᵢ** | Temps de service du client *i* | 10 min |

Nous définissions ensuite les variables de décision qui permettent de décrir les choix effectués par l’algorithme :  
quel est la capcité restante du camion lors de son déplacement, où, et à quel moment:

| Variable | Signification | Type |
|---|---|---|
| **xᵢⱼᵏ** | 1 si le camion *k* va directement de *i* à *j* | Binaire |
| **yᵢᵏ** | 1 si le client *i* est servi par le camion *k* | Binaire |
| **tᵢ** | Heure d’arrivée au point *i* | Réel |
| **qᵢᵏ** | Charge restante du camion *k* après le point *i* | Réel |

L’objectif principal est donc de **minimiser le coût total des tournées**, en tenant compte des trajets parcourus par tous les camions :

$$
\min \sum_{k \in K} \sum_{i \in V} \sum_{j \in V} c_{ij} \, x_{ijk}
$$

<br>

Nous ajoutons à cet objectif des contraintes qui assurent la cohérence, la faisabilité opérationnelle et le respect des règles logistiques.

Chaque client doit être visité **exactement une fois**, par un seul camion.

$$
\sum_{k \in K} \sum_{j \in V} x_{ij}^k = 1 \qquad \forall i \in V \setminus \{0\}
$$

Si le camion entre dans un point, il doit en ressortir.

$$
\sum_{j} x_{ij}^k = \sum_{j} x_{ji}^k \qquad \forall i, \forall k
$$

Chaque camion part du dépôt et y revient exactement une fois.

$$
\sum_{j} x_{0j}^k = 1,\qquad 
\sum_{i} x_{i0}^k = 1 \qquad \forall k
$$

La somme des demandes des clients desservis par un camion ne doit jamais dépasser sa capacité.

$$
\sum_{i \in V} d_i \, y_i^k \le Cm_k \qquad \forall k
$$

Le camion doit arriver dans la fenêtre autorisée.

$$
a_i \le t_i \le b_i
$$

Cohérence temporelle :

$$
t_j \ge t_i + c_{ij} + s_i - M(1 - x_{ij}^k)
$$

**M** est une *grande constante artificielle* utilisée dans certains modèles d’optimisation pour **activer ou désactiver une contrainte** en fonction d’une variable binaire.

Son rôle est de permettre des contraintes du type :

- Si le camion emprunte l’arc (i,j), alors la relation temporelle doit être respectée.
- Sinon, on désactive cette relation.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> Complexité du problème :</h3> 

La complexité algorithmique est une estimation du temps de calcul ou de la quantité de mémoire requise par un algorithme en fonction de la taille de ses données d’entrée, notée (n). Elle est utilisée pour prédire les performances d’un algorithme et si celui-ci est scalable ou non lorsque le nombre de données augmente.
Il y a deux types de complexité :

- Complexité de temps : c’est le nombre d’opérations élémentaires qui doivent être effectuées pour que l’algorithme se termine. Complexité en espace
- Mémoire requise pendant l’exécution: c'est l'espace nécessaire à l’exécution d’un algorithme
Pour mieux situer les TSP et VRP dans le paysage de la complexité algorithmique, il est utile de considèrer la distinction entre les principales classes de problème: P, NP, NP-complet, NP-difficile.
Ces classes peuvent être utilisées pour définir la difficulté et le temps de calcul nécessaire pour résoudre un problème ou pour tester sa solution.

<br>

| **Type de problème** | **Vérifiable en temps polynomial (P)** | **Résoluble en temps polynomial (P)** |
|------------------------|---------------------------------------|--------------------------------------|
| **P** | Oui | Oui |
| **NP** | Oui | Oui ou Non |
| **NP-Complet** | Oui | Inconnu |
| **NP-Difficile** | Oui ou Non | Inconnu |

<br>

- Un problème de type NP qui est aussi de type P est **résoluble en temps polynomial.**
- Un problème NP-difficile qui est également NP-Complet est **vérifiable en temps polynomial.**

### Complexité du problème du VRP

Le **Vehicle Routing Problem (VRP)** appartient à la catégorie des problèmes les plus complexes en optimisation combinatoire.  
On peut résumer sa classification ainsi :

- Le VRP est **NP-difficile** (*NP-hard*).  
- Même ses sous-problèmes simplifiés, comme le TSP (qui consiste à trouver un seul cycle optimal), sont déjà **NP-complets**.  
- Il n’existe **aucun algorithme connu** capable de résoudre le VRP en **temps polynomial** pour toutes les instances.

#### Justification de la complexité

Le VRP combine plusieurs dimensions difficiles :

1. **Dimension du TSP**  
   Chaque tournée effectuée par un camion est un **problème du voyageur de commerce**, qui est lui-même NP-complet.

2. **Assignation des clients aux véhicules**  
   Il faut déterminer *quel véhicule dessert quel client*, ce qui ajoute une combinatoire supplémentaire.

3. **Contraintes additionnelles**  
   Le VRP intègre des contraintes comme :
   - les capacités des camions,
   - les fenêtres de temps,
   - les temps de service,
   - les incompatibilités véhicule–marchandise,
   - les points de collecte avant livraison, etc.

   Chaque contrainte augmente la taille de l’espace des solutions.

4. **Explosion combinatoire**
   Pour $ n $ clients et $ k $ véhicules :
   - Même sans contraintes, le nombre de solutions possibles est supérieur à :
     $$
     k^n \times (n!) \quad \text{(ordre de grandeur)}
     $$
   - Ce qui est **exponentiel**.

#### Complexité temporelle
En pratique :

- Le VRP **ne peut pas être résolu en temps polynomial** (sauf cas très trivial).
- Les algorithmes exacts (Branch & Bound, Branch & Cut, MILP…) ont un **temps d’exécution exponentiel**, typiquement :
  $$
  O(n!)
  $$
  ou, dans le meilleur des cas optimisés :
  $$
  O(2^n)
  $$

C’est pour cela que :

**aucune méthode exacte n'est utilisable pour des instances de plus de 50–100 clients** dans des temps raisonnables.  
On utilise des **métaheuristiques** (comme l’ALNS), capables de produire des solutions proches de l’optimal **en temps polynomial**, mais **sans garantie d’optimalité**.

### Conclusion sur la classification du VRP

- Le VRP **n’est pas dans P** (aucun algorithme polynomial connu).  
- Le VRP n’est **pas NP-complet**, car il n’est même pas vérifiable en temps polynomial (certaines variantes incluent des ressources non trivialement vérifiables).  
- Le VRP est **NP-difficile (NP-hard)**, ce qui signifie :
  > Il est au moins aussi difficile que les problèmes les plus difficiles de NP.

C’est cette classification qui justifie :
- l’usage d’approches heuristiques et métaheuristiques,
- l’évaluation expérimentale,
- et la recherche d'un compromis trajectoire/qualité/temps.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> Choix et définition de la méthode :</h3> 

C’est une métaheuristique de type ruin & recreate (détruire / réparer) qui combine plusieurs opérateurs de destruction (ruin) et réparation (recreate).
L’adaptativité signifie que l’algorithme apprend au fil des itérations quels opérateurs sont performants et augmente leur probabilité de réutilisation.

Principe essentiel :

- Partir d’une solution initiale S.
- Choisir un opérateur de destruction R et un opérateur de réparation A ayant chacune une probabilité similaire d’être sélectionnée.
- Appliquer R(S), solution partielle S' (quelques clients retirés).
- Appliquer A(S'), solution candidate S_new.
- Évaluer S_new, appliquer une règle d’acceptation.
- Mettre à jour les poids/scores des opérateurs en fonction de l’issue.
- Répéter jusqu’à critère d’arrêt.

**S est la solution initiale**, l'on utilise une algorithmie gloutonne pour construire une solution initiale faisable.

**R sont les opérateurs de destruction** qui retirent un certain nombre de clients de la solution courante, il en existe plusieurs types
  * Random removal : retirer des clients aléatoirement.
  * Worst removal : retirer les clients contribuant le plus au coût.
  * Shaw removal : retirer des clients similaires (proximité géographique, demande, etc.).

**A sont les opérateurs de réparation** qui réinsèrent les clients retirés dans la solution partielle, il en existe aussi plusieurs types
  * Greedy insertion : insérer chaque client au meilleur emplacement possible.
  * Regret insertion : insérer les clients en fonction du regret de ne pas les insérer immédiatement.
  * Best insertion : comme greedy mais tri des clients par coût d’insertion.

**Mise à jour dynamique** : Après chaque itération, les performances des opérateurs R et A sont évaluées si ils ont contribué à l’amélioration de la solution et leurs poids sont ajustés pour favoriser ceux qui réussissent le mieux.

**Critère d’arrêt** : L'algorithme s’arrête aux au nombre maximum d'itérations, lorsque le temps fixé est écoulé, ou bien lorsque le meilleur objectif n’est plus amélioré.

L'algorithme ALNS s'avère être une méthode très prometteuse pour le traitement des problèmes complexes de tournée de véhicules. Sa puissance réside dans sa flexibilité et son adaptabilité : en combinant dynamiquement plusieurs opérateurs de destructions/réparations, il réalise une exploration intelligente de vastes zones dans l’espace de recherche tout en se basant sur les performances observées durant les itérations pour orienter sa recherche.

```python
print(f"Instance chargée : {instance['dimension']} nœuds, capacité {instance['capacity']}"

- `instance['dimension']` : indique le **nombre total de nœuds** dans l’instance.  
  Ce nombre inclut **tous les clients** ainsi que le **dépôt**.  
  Par exemple, si `dimension = 51`, cela signifie qu’il y a **50 clients + 1 dépôt**.

- `instance['capacity']` : représente la **capacité maximale du camion** de charge transportée.  
  Cette valeur limite la **somme des demandes des clients** visités sur une même route.


<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 2. Solution initiale :</h3>

La génération d'une solution initiale constitue le point de départ de l'algorithme ALNS. Nous avons implémenté une heuristique constructive simple mais efficace qui respecte les contraintes de capacité des véhicules.

**Principe de l'heuristique :**
- Partir du dépôt avec une charge nulle
- Ajouter consécutivement les clients à la route courante selon l'odre de numéro des clients
- Retourner au dépôt lorsque la capacité du véhicule est atteinte
- Créer une nouvelle route pour les clients restants

Voici un extrait du code présantant la fonction utilisée:

```python
def solution_initiale(instance):
    """
    Génère une solution initiale en utilisant une heuristique simple avec les données VRPLib.
    """
    depot = instance["depot"][0]  # Index du dépôt (généralement 0)
    demand = instance["demand"]   # Numpy array des demandes
    capacity = instance["capacity"]  # Capacité des véhicules
    n = instance["dimension"]     # Nombre de nœuds
    
    # Les clients sont tous les nœuds sauf le dépôt
    clients = []
    i = 0
    while i < n:
        if i != depot:
            clients.append(i)
        i += 1
 
    tour = [depot]
    p = 0
    demande = 0
    while p < len(clients):
        client_demande = demand[clients[p]]
        demande += client_demande
        if capacity is not None and demande > capacity:
            tour.append(depot)
            demande = client_demande  # Commencer nouvelle route avec ce client
            tour.append(clients[p])
            p += 1
        else:
            tour.append(clients[p])
            p += 1
    tour.append(depot)
    
    # Convertir le tour en routes
    routes = []
    courants = []
    i = 0
    n_tour = len(tour)
    while i < n_tour:
        v = tour[i]
        if v == depot:
            if len(courants) > 0:
                # Fermer la route: [depot] + clients + [depot]
                route = [depot]
                j = 0
                while j < len(courants):
                    route.append(courants[j])
                    j += 1
                route.append(depot)
                routes.append(route)
                courants = []
        else:
            courants.append(v)
        i += 1
    return routes
```

Cette approche assure une solution qui peut être utilisée comme base pour l’amélioration future de l’ALNS. Cette heuristique, bien que simple, produit des solutions admissibles et cohérentes, qui sont d’excellents points de départ pour l’optimisation. Dans notre cas, il y a plus de chances pour que l’algorithme converge vers une meilleure solution s’il réalise une recherche approfondie dans l’espace de solutions, ce qui est favorisé par une solution initiale sous-optimale. 
L'emploi d'un algorithme glouton aurait conduit à une solution trop fixe, empêchant d'enduire d'autres solutions des premiers résultats par exploration séquentielle sous forme d’ALNS. C'est la raison pour laquelle nous avons opté pour une heuristique simple triée par le numéro de client plutôt que la construction gloutonne.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 3. Opérateurs implémentés :</h3>

L'ALNS repose sur l'application itérative d'opérateurs de destruction et de réparation. Nous avons implémenté plusieurs opérateurs pour assurer la diversification et l'intensification de la recherche.

<h4 style="color:#1f4e79; font-style:italic; margin-top:0"> Opérateurs de destruction :</h4>

Ils retirent temporairement une partie de la solution (ex : un ensemble de clients) afin de "casser" la structure actuelle. Leur rôle principal est de **diversifier la recherche**, c’est-à-dire éviter que l’algorithme reste bloqué dans une solution locale. En supprimant certains éléments stratégiques, on libère de nouvelles possibilités de reconstruction. Il existe une dizaine d'opérateurs de destructions mais voici les 3 qui nous semblent pertinent dans notre contexte:

**Worst Removal :** Retire les clients impliqués dans les arêtes les plus coûteuses de la solution courante. Cet opérateur vise les parties de la solution les moins efficaces pour permettre leur reconstruction optimisée.

**Random Removal :** Retire aléatoirement un nombre fixé de clients. Cette approche favorise l’exploration, qui consiste à examiner des parties de l’espace des solutions qui ne sont pas ciblées par les heuristiques déterministes.

**Shaw Removal :** Retire des clients géographiquement proches. Cette stratégie exploite la structure spatiale du problème en supposant que des clients voisins peuvent être regroupés efficacement.

<h4 style="color:#1f4e79; font-style:italic; margin-top:0"> Opérateurs de réparation :</h4>

Ils reconstruisent la solution après la destruction en réinsérant les clients retirés, selon une logique plus ou moins optimisée. Leur rôle est d'**intensifier la recherche**, c’est-à-dire améliorer ou stabiliser la solution en choisissant des positions pertinentes. Il existe également une dizaine d'opérateurs de destructions mais voici les 3 qui nous semblent pertinent dans notre contexte:

**Greedy Insertion :** Réinsère chaque client retiré à la position qui minimise l'augmentation du coût total. Cette approche privilégie les décisions localement optimales.

**Best Insertion :** Pour chaque client à insérer, évalue toutes les positions possibles et choisit la meilleure. Cette méthode est plus coûteuse pour le temps d'executions mais produit souvent de meilleures solutions.

**Regret Insertion :** Calcule le "regret" de chaque client, défini comme la différence entre son meilleur coût d'insertion et son second meilleur.Les clients avec le plus grand regret sont insérés en priorité, évitant ainsi les décisions difficiles.

Ces deux types d’opérateurs fonctionnent ensemble : 
la destruction introduit du changementtandis que la réparation donne une structure cohérente à ce changement. C’est cette enchaînement qui permet à l’ALNS d’explorer efficacement l’espace des solutions tout en s'approchant vers des résultats de plus en plus performants.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 4. Critère d'acceptation :</h3>

Le critère d'acceptation détermine si une solution candidate remplace la solution courante. Nous avons implémenté le mécanisme du **Simulated Annealing** qui permet d'accepter des solutions dégradantes avec une probabilité qui diminue au cours du temps.

**Principe du Simulated Annealing :**
- Toute solution améliorante (Δ < 0) est automatiquement acceptée
- Une solution dégradante est acceptée avec une probabilité: $ P = e^{-\Delta / T} $
- La température T diminue progressivement selon la formule: $ T = α × T $

Voici un extrait de la fonction utilisée:

```python
def acceptation_regle(delta, T, mode="sa"):
    if delta < 0:
        return True  # Solution améliorante
    
    if mode == "sa":
        prob = math.exp(-delta / T)
        return random.random() < prob
    
    return False  # Refuser les solutions dégradantes
```

**Paramètres de refroidissement :**
- **Température initiale (T₀)** : Détermine l'acceptation initiale des solutions dégradantes
- **Facteur de refroidissement (α)** : Contrôle la vitesse de convergence (généralement α ∈ [0.99, 0.999])

Ce mécanisme permet de prévenir les minima locaux au début d'exécution et d'aboutir à un bon résultat à la fin du processus. Le compromis entre l'exploration et l'exploitation est ainsi naturellement contrôlé par la température.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 5. Mécanisme adaptatif :</h3>

Le mécanisme adaptatif représente la première innovation de l'ALNS : c'est l'apprentissage dynamique de la performance des opérateurs qui permet de guider la recherche vers les stratégies les plus performantes.

**Système de notation :**
Chaque opérateur gagne un score basé sur la qualité des solutions qu'il génère :
- **σ₁** : Nouvelle meilleure solution globale
- **σ₂** : Amélioration de la dernière solution générée
- **σ₃** : Solution acceptée mais non améliorante
- **σ₄** : Solution rejetée

**Mécanisme de pondération :**
Les poids des opérateurs évoluent selon une moyenne mobile pondérée :

Voici un extrait de code de la fonction utilisée:

```python
def selection_op(scores):
    # Tables de poids
    w_remove = {}
    for op, data in scores["remove"].items():
        w_remove[op] = data["weight"]

    w_repair = {}
    for op, data in scores["insert"].items():
        w_repair[op] = data["weight"]

    # Tirage par roulette (pondéré)

    #Générer une probabilité pour chaque opérateur de destruction à partir des poids
    total = sum(w_remove.values())
    proba_destruction = {}
    for k, v in w_remove.items():
        proba_destruction[k] = v / total


    # Tirer un opérateur de destruction selon la probabilité calculée
    op_destruct = random.choices(list(proba_destruction.keys()), weights=list(proba_destruction.values()))[0]

    # Générer une probabilité pour chaque opérateur de réparation à partir des poids
    total = sum(w_repair.values())
    proba_reparation = {}
    for k, v in w_repair.items():
        proba_reparation[k] = v / total

    # Tirer un opérateur de réparation selon la probabilité calculée
    op_repare = random.choices(list(proba_reparation.keys()), weights=list(proba_reparation.values()))[0]
    return op_destruct, op_repare
```

La sélection des opérateurs suit une **roulette pondérée** où les opérateurs performants ont une probabilité plus élevée d'être choisis :

$$ P(opérateur_i) = poids_i / Σ(poids_j) $$

Les scores sont remis à zéro tous les `segment_len` itérations pour maintenir la réactivité aux changements de contexte et éviter la sur-exploitation d'opérateurs temporairement efficaces.

Ce mécanisme permet à l'algorithme d'**apprendre en temps réel** quelles stratégies sont les plus adaptées au problème traité et d'ajuster automatiquement son comportement pour maximiser l'efficacité de la recherche.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 6. Fonction d'acceptation et mise à jour :</h3>

La fonction `appli_acceptation` représente le **cœur de l'algorithme ALNS** : elle orchestre l'ensemble du processus de décision d'acceptation/rejet d'une solution candidate et met à jour tous les paramètres associés.

Cette fonction centralise quatre opérations :
1. **Décision d'acceptation** via le critère choisi (Simulated Annealing)
2. **Mise à jour de l'état courant** si la solution est acceptée
3. **Gestion du meilleur global** pour conserver la meilleure solution trouvée
4. **Attribution des crédits aux opérateurs** selon la qualité du résultat

**Mécanisme de crédit adaptatif :**
Selon l'outcome de la solution candidate, les opérateurs utilisés reçoivent des scores différents :

```python
# Classification des résultats
if C_new < state["C_best"]:
    outcome = "best_global"      # score += 50.0 points
elif delta < 0:
    outcome = "improve"          # score += 25.0 points  
else:
    outcome = "accepted_worse"   # score += 3.0 points
# Rejet -> score += 0.0 point (pas de crédit)
```

**Mise à jour périodique des poids :**
Tous les `segment_len` itérations, les poids des opérateurs sont recalculés selon une **moyenne mobile pondérée** :

$$ poids_{nouveau} = (1 - ρ) × poids_{ancien} + ρ × \frac{score_{accumulé}}{utilisations} $$

où ρ ∈ [0,1] contrôle la vitesse d'adaptation (généralement ρ = 0.65).


Cette approche garantit ainsi la **cohérence** entre tous les mécanismes adaptatifs de l'ALNS et permet un apprentissage automatique des meilleures stratégies de recherche au fil des itérations.

### Implémentation de la 1ʳᵉ contrainte : les fenêtres de temps

Dans ce projet, nous considérons une variante du Problème de Tournée de Véhicules (VRP) avec **fenêtres de temps**. En plus de servir chaque client une seule fois et de respecter la capacité du véhicule, le livreur doit respecter, pour chaque client, un **intervalle de temps** pendant lequel la visite est autorisée. Cette contrainte modélise par exemple des horaires d’ouverture de magasins ou des rendez-vous imposés.

---

#### 1. Rappel théorique

Pour chaque client \(i\), on dispose :

- d’une **fenêtre de temps** \([e_i, l_i]\)  
  - \(e_i\) : heure la plus tôt à laquelle le véhicule est autorisé à commencer le service chez le client \(i\) ;  
  - \(l_i\) : heure limite à ne pas dépasser pour commencer le service ;
- d’une **durée de service** \(s_i\), correspondant au temps nécessaire pour charger/décharger, livrer, etc.

On note :

- \(t_{ij}\) : temps de trajet entre les clients \(i\) et \(j\) ;
- \(A_i\) : heure d’arrivée du véhicule au client \(i\).

Le véhicule peut arriver **avant** le début de la fenêtre \(e_i\) et attendre, mais une arrivée **après** \(l_i\) rend la tournée **invalide** (ou très pénalisée, selon la politique choisie).

Pour un arc \((i, j)\) de la tournée, avec un temps de trajet \(t_{ij}\), l’heure d’arrivée au client \(j\) est donnée par la relation de propagation suivante :

$$
A_j = \max\big(e_j,\; A_i + s_i + t_{ij}\big)
$$

- \(A_i + s_i + t_{ij}\) : heure à laquelle le véhicule **aurait** dû arriver chez \(j\) en partant directement de \(i\) après le service ;
- \(\max(e_j, \cdot)\) : si le véhicule arrive trop tôt, il **attend** jusqu’au début de la fenêtre \(e_j\).

La contrainte de fenêtre de temps est respectée si, pour tout client \(j\) de la tournée :

$$
A_j \le l_j \quad \text{pour tout client } j
$$

---

#### 2. Représentation des données dans le code

Dans notre implémentation, les fenêtres de temps et les durées de service sont stockées dans des tableaux (ou dictionnaires) indexés par le numéro de client :

- `earliest[i]` représente \(e_i\), le début de la fenêtre de temps du client \(i\) ;
- `latest[i]` représente \(l_i\), la fin de la fenêtre de temps du client \(i\) ;
- `service[i]` représente \(s_i\), la durée de service chez le client \(i\).

Les temps de trajet sont stockés dans une matrice :

- `travel_time[i][j]` représente \(t_{ij}\), le temps de trajet entre les clients \(i\) et \(j\).

Une solution (une tournée d’un véhicule) est représentée par une liste de clients, par exemple :

```python
route = [0, 5, 3, 7, 2, 0]  # 0 = dépôt


#### Extrait de code – Gestion des fenêtres de temps

Le code suivant illustre concrètement la manière dont les fenêtres de temps sont prises en compte dans notre algorithme de livraison avec fenêtres de temps.  
On y calcule l’heure d’arrivée à chaque client à partir des temps de trajet, des durées de service et des bornes \([e_i, l_i]\).  
Si l’arrivée chez un client dépasse la fin de sa fenêtre de temps, la tournée est considérée comme non faisable.


```python

# 0 = dépôt
depot = 0

# Fenêtres de temps (e_i et l_i) et durées de service s_i
# Index : 0 = dépôt, 1..4 = clients
earliest = [0, 10, 15, 5, 20]      # e_i : début de fenêtre
latest   = [999, 25, 35, 30, 40]   # l_i : fin de fenêtre
service  = [0, 5, 5, 5, 5]         # s_i : durée de service

# Matrice des temps de trajet t_ij (en minutes par exemple)
travel_time = [
    #  0   1   2   3   4   <-- vers j
    [  0,  5, 10,  8, 12],  
    [  5,  0,  4,  7,  6],  
    [ 10,  4,  0,  3,  5],  
    [  8,  7,  3,  0,  4],  
    [ 12,  6,  5,  4,  0],  
]


def respecte_fenetres_temps(route, travel_time, earliest, latest, service, depot=0):
    """
    Vérifie si une tournée respecte les fenêtres de temps.
    """
    current = route[0]
    time = earliest[current]  # heure de départ au dépôt

    for j in route[1:]:
        # trajet de current vers j
        time += travel_time[current][j]

        # si on arrive avant le début de la fenêtre, on attend
        if time < earliest[j]:
            time = earliest[j]

        # si on dépasse la fin de la fenêtre alors contrainte violée
        if time > latest[j]:
            return False

        # on ajoute la durée de service chez j
        time += service[j]
        current = j

    return True


# Exemple d'utilisation

route1 = [0, 1, 3, 2, 4, 0]   # tournée candidate

if respecte_fenetres_temps(route1, travel_time, earliest, latest, service, depot):
    print(" La tournée", route1, "respecte les fenêtres de temps.")
else:
    print(" La tournée", route1, "viole au moins une fenêtre de temps.")
```

<h1 style="color:#1f4e79; font-style:italic; margin-top:0"> Implémentation de la 2ème contrainte : Contrainte de trafic dynamique</h1>

Dans ce projet, nous avons abordé le problème de tournées de véhicules (Vehicle Routing Problem – VRP) avec une contrainte supplémentaire, la variation temporelle du trafic. L’idée était de reproduire plus fidèlement les conditions réelles de la circulation, la durée d’un trajet dépendant du moment de la journée (périodes de pointe et trafic normal). Cette méthode conduit à des tournées de livraison plus réelles et mieux optimales dans un environnement urbain.

**FONCTIONNEMENT DE LA CONTRAINTE DE TRAFIC DYNAMIQUE :**

Dans un VRP classique, le temps (ou le coût) entre deux clients $i$ et $j$ est constant et noté $T_{i,j}$.
Dans notre modèle ce cout devient dépendant du temps
$$ C_{i,j} = T_{i,j} * f(t) $$
où $f(t)$ représente un facteur de trafic qui varie selon l’heure de passage $t$
Par exemple, les trajets sont plus longs pendant les heures de pointe du matin et du soir.

La rélation définissant le facteur de trafic $f(t)$ provient de la formule **BPR (Bureau of Public Roads)**, un modèle empirique standard en ingénierie du trafic elle stipule:

$$ f(t) = 1 + α * ( \frac{V}{C} )^β $$

où :
- $V$ : volume de trafic (nombre de véhicules)
- $C$ : capacité de la route (nombre maximal de véhicules)
- $α$ et $β$ : paramètres empiriques (cas typiques : $α=0.15$ et $β=4$)

La valeur de la capacité $C$ est fixée en fonction du type de route (ex : route urbaine, autoroute). Étant que nous sommes dans notre VRP ce passe dans une zone urbaine l'on obtien une constante $C=3200 véh/h$ issues de calibrations du Cerema (France) et du Highway Capacity Manual (HCM).

Et enfin , le volume de trafic $V$ varie sélon l'heure de la journée dans une journée type en France selon les Données agrégées Sytadin, et Cerema Île-de-France (2023) : Nous avons un tableau de valeur pour chaque heure de la journée.

Cette robustesse est essentielle en contexte industriel : elle garantit que l'algorithme produira des solutions de qualité prévisible, indépendamment des variations aléatoires inhérentes à la méthode.

**Phases d'optimisation identifiées :**

1. **Phase d'amélioration rapide (0-100 itérations)** :
   - Chute dramatique du coût : d'environ 60 000 à 35 000 ce qui montre une amélioration de 42%

   - Les améliorations sont fréquentes car l'espace d'amélioration est vaste

2. **Phase d'exploration intensive (100-400 itérations)** :
   - Amélioration progressive mais soutenue : de d'environ entre 35 000 à 31 000
   - L'algorithme explore différentes régions de l'espace des solutions
   - La température élevée permet encore l'acceptation de solutions dégradantes temporaires

3. **Phase de convergence fine (400-1200 itérations)** :
   - Stabilisation autour de 30 000 avec des améliorations marginales
   - La température décroissante rend l'algorithme plus sélectif
   - Les dernières optimisations portent sur des ajustements locaux de haute qualité

**Mécanismes algorithmiques observés :**

- **Efficacité du recuit simulé** : La courbe montre une convergence progressive sans oscillations majeures, preuve que les paramètres de température sont bien ajustés
- **Impact de l'apprentissage adaptatif** : Les améliorations régulières suggèrent que le système de pondération des opérateurs identifie efficacement les stratégies performantes
- **Équilibre exploration/exploitation** : La transition fluide entre les phases montre un bon compromis entre diversification (début) et intensification (fin)

**Analyse de l'efficacité temporelle :**
Environ **80% de l'amélioration totale** est obtenue dans les **200 premières itérations** (soit ~20% du temps total), démontrant l'efficacité de l'approche pour obtenir rapidement des solutions de bonne qualité. Cette caractéristique est particulièrement valuable pour des applications avec contraintes de temps strictes.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Analyse des performances :</h3>

#### Pour petite instance (A-n32-k5, 32 clients)
L’algorithme atteint une solution **quasi optimale** avec un gap très faible (**0,39%**).  
La faible température initiale (80) et une destruction modérée (q_remove = 5) permettent une convergence rapide et robuste.  
Le temps de calcul est minimal (≈ 4 s), montrant l’efficacité de l’approche sur les petites tailles.

#### Pour instances moyennes (X-n101-k25 et X-n200-k36)
Pour des tailles plus complexes :

- le gap reste contenu (**3,98%** à **6,48%**),
- les temps restent raisonnables (environ 20 à 100 s),
- la température (100–120) facilite une exploration suffisamment large,
- q_remove de 7 à 9 offre un compromis entre destruction et stabilité des routes.

La progression du gap est cohérente avec la difficulté croissante : plus il y a de clients, plus l’espace de recherche explose.

#### pour très grande instance (X-n1001-k43)
L’instance à plus de **1000 clients** révèle les limites actuelles du modèle :

- le gap grimpe à **73%**, indiquant une forte difficulté à explorer un espace aussi vaste,
- la température élevée (150) et q_remove = 12 ne suffisent pas à compenser l’explosion combinatoire,
- l’algorithme atteint la limite de temps de 300 s.

Ce comportement est attendu : les métaheuristiques ont besoin d’ajustements plus sophistiqués (opérateurs spécialisés, rechargement adaptatif de la température, etc.) pour bien fonctionner sur des instances massives.

Voici une interprétation des différents paramètres:

- **Température (T)**  
  Plus elle est élevée, plus l’algorithme accepte des solutions moins bonnes au début, favorisant l’exploration.  
  Elle augmente avec la taille du problème pour mieux parcourir l’espace de recherche.

- **q_remove**  
  Nombre d’éléments retirés dans l’opérateur de destruction.  
  Une valeur trop faible empêche la diversification, une valeur trop grande casse les routes.  
  Les valeurs utilisées montrent une augmentation logique avec la taille des instances.

- **Itérations**  
  Représentent la limite maximale du nombre d’améliorations possibles.  
  En pratique, le `max_time` coupe souvent l’exécution avant d’atteindre ce nombre, surtout pour les grandes instances.

Ce résultat était attendu, car :

- l’espace de recherche explose avec 1000+ clients,  
- le nombre d’itérations réalisables en 300 s est limité,  
- l’ALNS nécessite une adaptation plus poussée pour les grands problèmes.

De manière générale, l’algorithme présente :

- **une excellente performance** sur les petites instances,  
- **une bonne robustesse** jusqu’à plus de 200 clients,  
- **une baisse significative de qualité** sur les très grandes instances, où le temps limite est insuffisant pour converger.

Ces résultats restent ainsi cohérents avec les performances attendues d’une méta-heuristique et marquent une base solide pour des améliorations futures.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Perspectives d’amélioration :</h3>

Plusieurs améliorations peuvent être envisagées pour renforcer les performances de l’algorithme, en particulier sur les instances de grande taille (≥ 1000 clients) :

- **Renforcement des opérateurs de recherche**
  - Ajouter de nouveaux opérateurs de destruction/réparation afin d’explorer plus efficacement l’espace des solutions.
  - Renforcer la sélection adaptative des opérateurs en fonction de leur dernière performance.

- **Régulation dynamique des paramètres**
  - Réguler automatiquement `q`, la température, ou le taux de refroidissement en fonction de la qualité de la solution en cours.
  - Adapter le rythme de mise à jour des poids (`rho`, `segment_len`) selon la stabilité du coût.
  - Mettre en place une stratégie d’auto-calibration pour limiter les réglages manuels selon la taille de l’instance.

- **Prise en compte de contraintes supplémentaires**
  L’ajout de contraintes, comme une compatibilité entre véhicule et client et une limite de durée de tournée, rendrait l’algorithme applicable à des scénarios industriels plus réalistes.

- **Évaluation de la robustesse**
  - Réaliser des séries de tests sur des instances très larges (2000 clients) afin de valider la stabilité de l'algorithme.
  - Comparer les performances avec d’autres métaheuristiques de référence (recherche tabou, ACO, algorithme génétique, recuit simulé).
