# II.&nbsp;&nbsp; Modélisation

<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Rappel du problème formel :</h3>


On cherche à déterminer les **tournées optimales de livraison** pour un ensemble de camions partant d’un dépôt, visitant un ensemble de clients, puis revenant au dépôt.  
L’objectif est de **minimiser le coût total** (distance, durée ou énergie), tout en respectant les contraintes opérationnelles de capacités avec un ajout supplémentairede contrainte de fenêtre de temps et de trafic

Nous allons d'abord définir les paramètrese qui regroupent l’ensemble des données connues **avant la résolution** du problème.  
Ils définissent la structure du réseau, les caractéristiques des clients et les propriétés des véhicules:

| Nom | Signification | Exemple |
|---|---|---|
| **V** | Ensemble des clients + dépôt (0 = dépôt) | V = {0,1,2,3} |
| **K** | Ensemble des camions | K = {1,2} |
| **cᵢⱼ** | Coût entre *i* et *j* | c₁₂ = 12|
| **dᵢ** | Demande du client *i* | d₂ = 50|
| **Cmₖ** | Capacité du camion *k* | Cm₁ = 200|
| **[aᵢ, bᵢ]** | Fenêtre temporelle de service | [8h, 12h] |
| **sᵢ** | Temps de service du client *i* | 10 min |

Nous définissions ensuite les variables de décision qui permettent de décrir les choix effectués par l’algorithme :  
quel est la capcité restante du camion lors de son déplacement, où, et à quel moment:

| Variable | Signification | Type |
|---|---|---|
| **xᵢⱼᵏ** | 1 si le camion *k* va directement de *i* à *j* | Binaire |
| **yᵢᵏ** | 1 si le client *i* est servi par le camion *k* | Binaire |
| **tᵢ** | Heure d’arrivée au point *i* | Réel |
| **qᵢᵏ** | Charge restante du camion *k* après le point *i* | Réel |

L’objectif principal est donc de **minimiser le coût total des tournées**, en tenant compte des trajets parcourus par tous les camions :

$$
\min \sum_{k \in K} \sum_{i \in V} \sum_{j \in V} c_{ij} \, x_{ijk}
$$

<br>

Nous ajoutons à cet objectif des contraintes qui assurent la cohérence, la faisabilité opérationnelle et le respect des règles logistiques.

Chaque client doit être visité **exactement une fois**, par un seul camion.

$$
\sum_{k \in K} \sum_{j \in V} x_{ij}^k = 1 \qquad \forall i \in V \setminus \{0\}
$$

Si le camion entre dans un point, il doit en ressortir.

$$
\sum_{j} x_{ij}^k = \sum_{j} x_{ji}^k \qquad \forall i, \forall k
$$

Chaque camion part du dépôt et y revient exactement une fois.

$$
\sum_{j} x_{0j}^k = 1,\qquad 
\sum_{i} x_{i0}^k = 1 \qquad \forall k
$$

La somme des demandes des clients desservis par un camion ne doit jamais dépasser sa capacité.

$$
\sum_{i \in V} d_i \, y_i^k \le Cm_k \qquad \forall k
$$

Le camion doit arriver dans la fenêtre autorisée.

$$
a_i \le t_i \le b_i
$$

Cohérence temporelle :

$$
t_j \ge t_i + c_{ij} + s_i - M(1 - x_{ij}^k)
$$

**M** est une *grande constante artificielle* utilisée dans certains modèles d’optimisation pour **activer ou désactiver une contrainte** en fonction d’une variable binaire.

Son rôle est de permettre des contraintes du type :

- Si le camion emprunte l’arc (i,j), alors la relation temporelle doit être respectée.
- Sinon, on désactive cette relation.

```python
print(f"Instance chargée : {instance['dimension']} nœuds, capacité {instance['capacity']}"

- `instance['dimension']` : indique le **nombre total de nœuds** dans l’instance.  
  Ce nombre inclut **tous les clients** ainsi que le **dépôt**.  
  Par exemple, si `dimension = 51`, cela signifie qu’il y a **50 clients + 1 dépôt**.

- `instance['capacity']` : représente la **capacité maximale du camion** de charge transportée.  
  Cette valeur limite la **somme des demandes des clients** visités sur une même route.


<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 2. Solution initiale :</h3>

La génération d'une solution initiale constitue le point de départ de l'algorithme ALNS. Nous avons implémenté une heuristique constructive simple mais efficace qui respecte les contraintes de capacité des véhicules.

**Principe de l'heuristique :**
- Partir du dépôt avec une charge nulle
- Ajouter consécutivement les clients à la route courante selon l'odre de numéro des clients
- Retourner au dépôt lorsque la capacité du véhicule est atteinte
- Créer une nouvelle route pour les clients restants

Voici un extrait du code présantant la fonction utilisée:

```python
def solution_initiale(instance):
    """
    Génère une solution initiale en utilisant une heuristique simple avec les données VRPLib.
    """
    depot = instance["depot"][0]  # Index du dépôt (généralement 0)
    demand = instance["demand"]   # Numpy array des demandes
    capacity = instance["capacity"]  # Capacité des véhicules
    n = instance["dimension"]     # Nombre de nœuds
    
    # Les clients sont tous les nœuds sauf le dépôt
    clients = []
    i = 0
    while i < n:
        if i != depot:
            clients.append(i)
        i += 1
 
    tour = [depot]
    p = 0
    demande = 0
    while p < len(clients):
        client_demande = demand[clients[p]]
        demande += client_demande
        if capacity is not None and demande > capacity:
            tour.append(depot)
            demande = client_demande  # Commencer nouvelle route avec ce client
            tour.append(clients[p])
            p += 1
        else:
            tour.append(clients[p])
            p += 1
    tour.append(depot)
    
    # Convertir le tour en routes
    routes = []
    courants = []
    i = 0
    n_tour = len(tour)
    while i < n_tour:
        v = tour[i]
        if v == depot:
            if len(courants) > 0:
                # Fermer la route: [depot] + clients + [depot]
                route = [depot]
                j = 0
                while j < len(courants):
                    route.append(courants[j])
                    j += 1
                route.append(depot)
                routes.append(route)
                courants = []
        else:
            courants.append(v)
        i += 1
    return routes
```

Cette approche assure une solution qui sert de base pour les améliorations futurs de l’ALNS. Bien que simple, cette heuristique produit des solutions faisables et cohérentes qui constituent un excellent point de départ pour l’optimisation. 
Dans notre cas, l’algorithme s’adapte mieux lorsqu’il peut explorer largement l’espace des solutions, ce qui est favorisé par une solution initiale non optimisée. L’utilisation d’un algorithme glouton aurait conduit à une solution déjà trop structurée, limitant la capacité de l’ALNS à diversifier ses recherches et à trouver de meilleures solutions au fil des itérations. C’est pour cette raison que nous avons retenu une heuristique simple ordonnée par le numéro de client plutôt qu’une construction gloutonne.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 3. Opérateurs implémentés :</h3>

L'ALNS repose sur l'application itérative d'opérateurs de destruction et de réparation. Nous avons implémenté plusieurs opérateurs pour assurer la diversification et l'intensification de la recherche.

<h4 style="color:#1f4e79; font-style:italic; margin-top:0"> Opérateurs de destruction :</h4>

Ils retirent temporairement une partie de la solution (ex : un ensemble de clients) afin de "casser" la structure actuelle. Leur rôle principal est de **diversifier la recherche**, c’est-à-dire éviter que l’algorithme reste bloqué dans une solution locale. En supprimant certains éléments stratégiques, on libère de nouvelles possibilités de reconstruction. Il existe une dizaine d'opérateurs de destructions mais voici les 3 qui nous semblent pertinent dans notre contexte:

**Worst Removal :** Retire les clients impliqués dans les arêtes les plus coûteuses de la solution courante. Cet opérateur vise les parties de la solution les moins efficaces pour permettre leur reconstruction optimisée.

**Random Removal :** Retire aléatoirement un nombre fixé de clients. Cette approche favorise l’exploration, qui consiste à examiner des parties de l’espace des solutions qui ne sont pas ciblées par les heuristiques déterministes.

**Shaw Removal :** Retire des clients géographiquement proches. Cette stratégie exploite la structure spatiale du problème en supposant que des clients voisins peuvent être regroupés efficacement.

<h4 style="color:#1f4e79; font-style:italic; margin-top:0"> Opérateurs de réparation :</h4>

Ils reconstruisent la solution après la destruction en réinsérant les clients retirés, selon une logique plus ou moins optimisée. Leur rôle est d'**intensifier la recherche**, c’est-à-dire améliorer ou stabiliser la solution en choisissant des positions pertinentes. Il existe également une dizaine d'opérateurs de destructions mais voici les 3 qui nous semblent pertinent dans notre contexte:

**Greedy Insertion :** Réinsère chaque client retiré à la position qui minimise l'augmentation du coût total. Cette approche privilégie les décisions localement optimales.

**Best Insertion :** Pour chaque client à insérer, évalue toutes les positions possibles et choisit la meilleure. Cette méthode est plus coûteuse pour le temps d'executions mais produit souvent de meilleures solutions.

**Regret Insertion :** Calcule le "regret" de chaque client, défini comme la différence entre son meilleur coût d'insertion et son second meilleur.Les clients avec le plus grand regret sont insérés en priorité, évitant ainsi les décisions difficiles.

Ces deux types d’opérateurs fonctionnent ensemble : 
la destruction introduit du changementtandis que la réparation donne une structure cohérente à ce changement. C’est cette enchaînement qui permet à l’ALNS d’explorer efficacement l’espace des solutions tout en s'approchant vers des résultats de plus en plus performants.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 4. Critère d'acceptation :</h3>

Le critère d'acceptation détermine si une solution candidate remplace la solution courante. Nous avons implémenté le mécanisme du **Simulated Annealing** qui permet d'accepter des solutions dégradantes avec une probabilité qui diminue au cours du temps.

**Principe du Simulated Annealing :**
- Toute solution améliorante (Δ < 0) est automatiquement acceptée
- Une solution dégradante est acceptée avec une probabilité: $ P = e^{-\Delta / T} $
- La température T diminue progressivement selon la formule: $ T = α × T $

Voici un extrait de la fonction utilisée:

```python
def acceptation_regle(delta, T, mode="sa"):
    if delta < 0:
        return True  # Solution améliorante
    
    if mode == "sa":
        prob = math.exp(-delta / T)
        return random.random() < prob
    
    return False  # Refuser les solutions dégradantes
```

**Paramètres de refroidissement :**
- **Température initiale (T₀)** : Détermine l'acceptation initiale des solutions dégradantes
- **Facteur de refroidissement (α)** : Contrôle la vitesse de convergence (généralement α ∈ [0.99, 0.999])

Ce mécanisme permet de prévenir les minima locaux au début d'exécution et d'aboutir à un bon résultat à la fin du processus. Le compromis entre l'exploration et l'exploitation est ainsi naturellement contrôlé par la température.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 5. Mécanisme adaptatif :</h3>

Le mécanisme adaptatif représente la première innovation de l'ALNS : c'est l'apprentissage dynamique de la performance des opérateurs qui permet de guider la recherche vers les stratégies les plus performantes.

**Système de notation :**
Chaque opérateur gagne un score basé sur la qualité des solutions qu'il génère :
- **σ₁** : Nouvelle meilleure solution globale
- **σ₂** : Amélioration de la dernière solution générée
- **σ₃** : Solution acceptée mais non améliorante
- **σ₄** : Solution rejetée

**Mécanisme de pondération :**
Les poids des opérateurs évoluent selon une moyenne mobile pondérée :

Voici un extrait de code de la fonction utilisée:

```python
def selection_op(scores):
    # Tables de poids
    w_remove = {}
    for op, data in scores["remove"].items():
        w_remove[op] = data["weight"]

    w_repair = {}
    for op, data in scores["insert"].items():
        w_repair[op] = data["weight"]

    # Tirage par roulette (pondéré)

    #Générer une probabilité pour chaque opérateur de destruction à partir des poids
    total = sum(w_remove.values())
    proba_destruction = {}
    for k, v in w_remove.items():
        proba_destruction[k] = v / total


    # Tirer un opérateur de destruction selon la probabilité calculée
    op_destruct = random.choices(list(proba_destruction.keys()), weights=list(proba_destruction.values()))[0]

    # Générer une probabilité pour chaque opérateur de réparation à partir des poids
    total = sum(w_repair.values())
    proba_reparation = {}
    for k, v in w_repair.items():
        proba_reparation[k] = v / total

    # Tirer un opérateur de réparation selon la probabilité calculée
    op_repare = random.choices(list(proba_reparation.keys()), weights=list(proba_reparation.values()))[0]
    return op_destruct, op_repare
```

La sélection des opérateurs suit une **roulette pondérée** où les opérateurs performants ont une probabilité plus élevée d'être choisis :

$$ P(opérateur_i) = poids_i / Σ(poids_j) $$

Les scores sont remis à zéro tous les `segment_len` itérations pour maintenir la réactivité aux changements de contexte et éviter la sur-exploitation d'opérateurs temporairement efficaces.

Ce mécanisme permet à l'algorithme d'**apprendre en temps réel** quelles stratégies sont les plus adaptées au problème traité et d'ajuster automatiquement son comportement pour maximiser l'efficacité de la recherche.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> 6. Fonction d'acceptation et mise à jour :</h3>

La fonction `appli_acceptation` représente le **cœur de l'algorithme ALNS** : elle orchestre l'ensemble du processus de décision d'acceptation/rejet d'une solution candidate et met à jour tous les paramètres associés.

Cette fonction centralise quatre opérations :
1. **Décision d'acceptation** via le critère choisi (Simulated Annealing)
2. **Mise à jour de l'état courant** si la solution est acceptée
3. **Gestion du meilleur global** pour conserver la meilleure solution trouvée
4. **Attribution des crédits aux opérateurs** selon la qualité du résultat

**Mécanisme de crédit adaptatif :**
Selon l'outcome de la solution candidate, les opérateurs utilisés reçoivent des scores différents :

```python
# Classification des résultats
if C_new < state["C_best"]:
    outcome = "best_global"      # score += 50.0 points
elif delta < 0:
    outcome = "improve"          # score += 25.0 points  
else:
    outcome = "accepted_worse"   # score += 3.0 points
# Rejet -> score += 0.0 point (pas de crédit)
```

**Mise à jour périodique des poids :**
Tous les `segment_len` itérations, les poids des opérateurs sont recalculés selon une **moyenne mobile pondérée** :

$$ poids_{nouveau} = (1 - ρ) × poids_{ancien} + ρ × \frac{score_{accumulé}}{utilisations} $$

où ρ ∈ [0,1] contrôle la vitesse d'adaptation (généralement ρ = 0.65).


Cette approche garantit ainsi la **cohérence** entre tous les mécanismes adaptatifs de l'ALNS et permet un apprentissage automatique des meilleures stratégies de recherche au fil des itérations.

  pi = {

### Résultats expérimentaux :

Les performances de l’algorithme ont été évaluées sur plusieurs instances de référence issues de la bibliothèque **VRPLIB**, couvrant différentes tailles de problèmes : petit (32 clients), moyen (101–200 clients) et très grand (1000+ clients).  
Ces tests permettent d’observer à la fois la qualité des solutions produites et la capacité de l’algorithme à s’adapter à l’augmentation de la complexité.

L’ensemble des résultats obtenus est présenté dans le tableau suivant :

| Instance | Coût optimal | Votre coût | Gap (%) | Temps (s) |
|-----------|--------------|-------------|----------|------------|
| A-n32-k5  | 784          | 787         | 0.39%    | 10.38      |

| X-n200-k36  | 58578      | 62209       | 6.2%     | 57.12      |

### Analyse des résultats

**• Petite instance (A-n32-k5)**  
Sur la plus petite instance, l’algorithme atteint une solution presque optimale.  
Le gap est très faible (**0,39%**), ce qui témoigne d’une excellente capacité d’exploration et d’un bon équilibre entre diversification et intensification.  
Le temps de calcul reste réduit (~10 s), montrant une très bonne efficacité sur les petites tailles.

**• Instances moyennes (X-n101-k25 et X-n200-k36)**  
Pour les instances plus complexes, le gap augmente légèrement, passant de **4,2%** à **6,2%**.  
Ces performances restent très satisfaisantes dans un cadre heuristique, surtout compte tenu de la croissance rapide de l’espace de recherche.  
Le temps de calcul demeure raisonnable (~30–60 s), ce qui confirme la bonne scalabilité de l’algorithme.

**• Très grande instance (X-n1001-k43)**  
Cette instance à plus de 1000 clients met en évidence les limites actuelles de l’approche.  
Avec un gap de **73,71%**, la solution obtenue s’éloigne nettement de l’optimal.  
Le temps d’exécution atteint la limite imposée (**300 secondes**), empêchant l’algorithme d’explorer suffisamment le voisinage.

Ce résultat était attendu, car :

- l’espace de recherche explose avec 1000+ clients,  
- le nombre d’itérations réalisables en 300 s est limité,  
- l’ALNS nécessite une adaptation plus poussée pour les très grands problèmes.

### Bilan global

Dans l’ensemble, l’algorithme présente :

- **une excellente performance** sur les petites instances,  
- **une bonne robustesse** jusqu’à ~200 clients,  
- **une baisse significative de qualité** sur les très grandes instances, où le temps limite est insuffisant pour converger.

Ces résultats restent cohérents avec les performances attendues d’une méta-heuristique générique et constituent une base solide pour des améliorations futures.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Perspectives d’amélioration :</h3>

Plusieurs améliorations peuvent être envisagées pour renforcer les performances de l’algorithme, en particulier sur les instances de grande taille (≥ 1000 clients) :

- **Renforcement des opérateurs de recherche**

- **Prise en compte de contraintes supplémentaires**
  L’ajout de contraintes, comme une compatibilité entre véhicule et client et une limite de durée de tournée, rendrait l’algorithme applicable à des scénarios industriels plus réalistes.

- **Évaluation de la robustesse**
  - Réaliser des séries de tests sur des instances très larges (2000 clients) afin de valider la stabilité de l'algorithme.
