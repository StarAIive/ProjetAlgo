# III.&nbsp;&nbsp;Représentation formelle
On cherche à déterminer les tournées optimales de livraison pour un ensemble de camions partant d’un dépôt, visitant un ensemble de clients, et revenant au dépôt.
L’objectif est de minimiser la durée totale de tournée ou la date de retour du dernier camion, en respectant certaines contraintes (fenêtres de temps, capacités, compatibilités, etc.).
<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Paramètres :</h3>

Cette partie présente l’ensemble des informations connues avant la résolution.
Elle regroupe les différentes données disponibles qui serviront à la modélisation et à la résolution du problème.
Ces éléments représentent donc les données d’entrée du modèle. Nous allons représenter les différentes données reconnues dans le tableau suivant:

| Nom | Signification | Exemple |
|:----|:---------------|:---------|
| **V** | Liste des villes ou clients à livrer (le dépôt est la ville de départ) | V = {0, 1, 2, 3} |
| **K** | Nombre de camions disponibles | K = {1, 2} |
| **cᵢⱼ** | Temps entre la ville *i* et la ville *j* | c₁₂ = 12 min |
| **dᵢⱼ** | Distance entre la ville *i* et la ville *j* | d₅ = 5km |
| **Qᵢ** | Quantité à livrer au client *i* | Q₂ = 50 kg |
| **Cmₖ** | Capacité maximale du camion *k* | Cm₁ = 200 kg |
| **[aᵢ, bᵢ]** | Heure d’ouverture et de fermeture pour livrer le client *i* | [8h, 12h] |
| **tᵢ** | Temps nécessaire pour effectuer la livraison chez le client *i* | t₃ = 10 min |
| **type(i)** | Type de marchandise à livrer | type(2) = “fragile” |
| **comp(k)** | Type de marchandises que le camion *k* peut transporter | comp(1) = {“normal”, “fragile”} |

Les variables de décision permettent de formaliser les choix à effectuer pour résoudre le problème.
Elles décrivent le comportement des camions, les trajets empruntés et les heures de livraison. Ces varaibles sont nécessaires pour construire une tournée optimale.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> Objectif et contraintes :</h3>

L’objectif de notre modèle est de déterminer les tournées de livraison permettant de **minimiser le coût total** ou la **durée totale des trajets**, tout en respectant les contraintes liées aux capacités des camions et aux fenêtres de temps.  

Dans le cas où plusieurs camions sont disponibles, il est également possible de chercher à **minimiser la date de retour du dernier camion**, afin d’équilibrer au mieux la charge de travail entre les véhicules.  

Les **contraintes** du modèle garantissent la cohérence et la faisabilité des tournées obtenues.  
Elles permettent de s’assurer que les livraisons respectent la logique du réseau, les capacités physiques et les horaires imposés.  

Voici les principales contraintes retenues :  

- **Contrainte de visite unique :** chaque client doit être livré **exactement une fois**, et par un seul camion.  
- **Contrainte de départ et de retour :** chaque camion **part du dépôt** au début de sa tournée et **y revient** à la fin.  
- **Contrainte de capacité :** la quantité totale transportée par un camion **ne doit jamais dépasser sa capacité maximale**.  
- **Contrainte de fenêtres de temps :** chaque livraison doit être effectuée **à l’intérieur de la plage horaire autorisée**.  
- **Contrainte de cohérence temporelle :** le temps d’arrivée au client suivant tient compte du **temps de trajet et du service** réalisé chez le client précédent.  
- **Contrainte de compatibilité :** certains camions ne peuvent livrer que certains types de marchandises.  

Ces différents points assurent que la solution obtenue soit réaliste, optimisée et directement exploitable dans ce contexte étudié.

# IV.&nbsp;&nbsp;Justification de la complexité
Pour finir, nous allons évaluer la complexité théorique.
Cette analyse permet de comprendre s’il est possible de résoudre le problème efficacement à l’aide d’un algorithme optimal, ou si celui-ci appartient à une classe de problèmes plus difficiles.
On cherche ainsi à savoir si le problème peut être résolu en temps polynomial, ou s’il est NP-difficile.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Mise en relation :</h3>

Le problème de gestion de tournées de livraison étudié ici peut être rapproché du **TSP (Travelling Salesman Problem)**.

Dans le TSP, un voyageur doit visiter une liste de villes **une seule fois**, puis **revenir à son point de départ**, tout en **minimisant la distance totale parcourue**.  
Ce problème est connu pour être **NP-difficile**, c’est-à-dire qu’aucun algorithme connu ne permet de le résoudre efficacement lorsque le nombre de villes devient important.

Notre problème de **tournées de livraison** aussi appelé **VRP (Vehicle Routing Problem)** est une **généralisation du TSP**, car il introduit des contraintes supplémentaires :

- Plusieurs **camions** au lieu d’un seul voyageur  
- Des **capacités limitées** pour chaque véhicule  
- Des **fenêtres de temps** pour les livraisons  
- Des **compatibilités camion/objet** à respecter  
- Une **variation du trafic** selon les heures de la journée  

Le TSP est donc bien inclus dans le VRP en amplifiant la difficulté du problème initial.
La **complexité** du TSP est donc **exponentielle**, notée en général **O(2ⁿ)** 
où *n* représente le nombre de villes à visiter.
 
Cela signifie que le temps de calcul **double pratiquement à chaque ajout de nouvelle ville**, rendant la résolution exacte rapidement impossible pour de grandes instances.

Puisque le TSP est **NP-difficile**, et que notre problème contient le TSP, on en déduit que le **VRP** ainsi que toutes ses variantes est lui aussi **NP-difficile**.  
En outre, **si nous savions résoudre notre problème rapidement**, nous saurions aussi résoudre le TSP ce qui est considéré actuellement comme impossible.

Par conséquent, notre problème appartient à la **classe des problèmes NP-difficiles**, ce qui justifie la recherche de **méthodes approchées** plutôt que de solutions exactes pour les grandes instances.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> Conséquences :</h3>

Le fait que notre problème appartienne à la **classe NP-difficile** a plusieurs conséquences importantes pour la résolution pratique.  

D’abord, cela signifie qu’il **n’existe pas de méthode exacte connue** capable de résoudre efficacement ce type de problème lorsque le nombre de villes ou de camions augmente.  
Les algorithmes exacts deviennent rapidement inexploitables, car leur temps de calcul croît **de manière exponentielle** avec la taille de l’instance.

Ainsi, pour des cas réels comportant **des dizaines ou centaines de points de livraison**, il devient nécessaire d’utiliser des **approches heuristiques**.  
Ces méthodes ne garantissent pas d’obtenir **la solution optimale**, mais elles permettent d’obtenir **une solution de bonne qualité** en un **temps de calcul raisonnable**.
