# I.&nbsp;&nbsp;Introduction
Cette introduction présente le cadre général du projet ainsi que les **bases de notre étude**.  
Il s’agit de déterminer en quoi la **recherche opérationnelle** et ses **outils d’optimisation combinatoire** peuvent contribuer à une **gestion intelligente** (*intelligent management*) des tournées de livraison, dans une perspective **écologique et économique**.  
 

<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Contexte :</h3>

Depuis les années 1990, la planète fait face à une problématique majeure :  
l' **augmentation de la consommation énergétique** et des **émissions de gaz à effet de serre (GES)**.  

Le secteur des transports est l’un des plus polluants à l’échelle mondiale.  
Les gouvernements, à l’image de la France, qui s’est fixé pour objectif une **division par 4 de ses émissions d’ici 2050**, cherchent notamment à :  
- Diminuer la pollution liée aux transports,  
- Améliorer les itinéraires et la logistique,  
- Promouvoir des technologies plus propres et plus efficaces.  

<br>

Dans ce contexte, **l’ADEME (Agence de la Transition Écologique)** redonne un élan fort à de nombreux projets novateurs en matière de **mobilité durable et intelligente**.  
L’objectif est de concevoir des systèmes de transport :  
- plus **écologiques**,  
- plus **efficaces**,  
- et mieux **adaptés grâce aux outils numériques** et à la **recherche opérationnelle**.  

<br>

L’entreprise **CesiCDP**, experte en mobilité multimodale intelligente, souhaite répondre à un appel à projets lancé par l’ADEME.  
Elle vise à démontrer que les **outils d’optimisation algorithmique** peuvent être mis au service d’une logistique plus durable, en permettant de :  
- programmer des tournées de livraison plus vertueuses,  
- réduire la distance totale parcourue,  
- et limiter les conséquences environnementales du transport de marchandises.  

<br>

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> Problématique :</h3>

Le projet cherche ainsi à répondre à la question suivante :  

> Comment planifier les itinéraires d’un camion devant livrer des colis dans plusieurs villes, afin de minimiser la distance parcourue, le temps total de tournée et, par conséquent, les émissions polluantes ?s

Autrement dit :  
- Concevoir un **algorithme efficace**, issu de la recherche opérationnelle,  
- Capable de **traiter des réseaux routiers de grande taille** (plusieurs milliers de points),  
- Et pouvant **intégrer des contraintes complexes** afin de refléter la réalité des transports modernes.

# II.&nbsp;&nbsp;Proprietes théoriques
On cherche à déterminer les **tournées optimales de livraison** pour un ensemble de camions partant d’un dépôt, visitant un ensemble de clients, et revenant au dépôt.  
L’objectif est de **minimiser les coûts totale de tournée**, tout en respectant certaines contraintes.  

<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Modélisation :</h3>

Il s’agit tout d’abord de **modéliser le réseau routier sous forme de graphe complet**, défini ainsi :  

$$
G = (V, E)
$$

où :  
- \( V = \{v1, v2, v3, ..., vn\} \) représente l’ensemble des villes ou points de livraison,  
- \( E \) représente l’ensemble des **arêtes** (routes) reliant ces villes,  
- chaque arête \((i, j)\) est associée à un **coût** \( c{ij} \) (distance, durée, consommation, etc.).  

<br>

**Représentation graphique :**

Le schéma suivant illustre une instance simple du problème.  
Chaque sommet correspond à une ville ou un client, et le sommet noir représente le **dépôt principal**.

Dans cette première étape, seuls les **sommets** sont représentés, c’est-à-dire les différents points de livraison.  

<br>

En ajoutant les arêtes reliant chaque ville, on obtient un **graphe** où chaque route correspond à un trajet possible entre deux points.  
À chaque arête \((i, j)\) est associé un **coût** \( c_{ij} \) représentant la **distance**, le **temps de parcours** ou la **consommation énergétique**.

<div style="text-align: center;"><img src="image/graphe2.png" width="300"></div>

Pour notre modélisation, le graphe \( G = (V, E) \) possède les caractéristiques suivantes :  

1. **Graphe non orienté :** les trajets entre deux villes sont considérés identiques dans les deux sens.  
2. **Graphe pondéré :** chaque arête \((i, j)\) est associée à un coût \( c_{ij} \) (distance, temps ou consommation).  
3. **Graphe connexe :** toutes les villes sont reliées directement ou indirectement, garantissant l’existence d’au moins une tournée valide.

Ce graphe représente la **base de notre modélisation mathématique** du problème de tournée de livraison.  
Il servira à formaliser les contraintes et l’objectif d’optimisation.  

<br>

Sous contrainte que **chaque sommet soit visité exactement une fois**.  

À noter que cet objectif est **très proche du Problème du Voyageur de Commerce (TSP)**, dans sa version d’optimisation.  
Nous pourrons donc nous appuyer sur ce modèle de référence pour construire et adapter notre propre solution.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> Complexité théorique :</h3> 

La **complexité algorithmique** désigne la mesure du **temps d’exécution** ou de la **mémoire utilisée** par un algorithme en fonction de la **taille des données d’entrée** notée \( n \).  
Elle permet d’évaluer la **performance** d’un algorithme et d’anticiper sa **scalabilité** lorsque le volume de données augmente.  

Il existe deux types de complexité :  
- **Complexité en temps :** mesure le nombre d’opérations nécessaires à l’exécution de l’algorithme.  
- **Complexité en espace :** mesure la quantité de mémoire requise pendant son exécution.

On y distingue plusieurs catégories :  
- Les **complexités logarithmiques** `O(log(n))` et **linéaires** `O(n)` sont très efficaces.  
- Les **complexités quadratiques** ou **polynomiales** (`O(n²)`, `O(n³)`) deviennent coûteuses pour de grands volumes de données.  
- Les **complexités exponentielles** (`O(2ⁿ)`) et **factorielles** (`O(n!)`) sont impraticables au-delà d’un certain seuil.

### Types de problèmes :

Afin de mieux comprendre la **position du TSP et du VRP** dans le domaine de la complexité algorithmique, il est utile de rappeler la distinction entre les principales classes de problèmes : **P, NP, NP-complets et NP-difficiles**.  

Ces classes permettent de caractériser la **difficulté de résolution** et la **vérifiabilité** d’un problème selon le temps de calcul nécessaire.

<br>

| **Type de problème** | **Vérifiable en temps polynomial (P)** | **Résoluble en temps polynomial (P)** |
|------------------------|---------------------------------------|--------------------------------------|
| **P** | Oui | Oui |
| **NP** | Oui | Oui ou Non |
| **NP-Complet** | Oui | Inconnu |
| **NP-Difficile** | Oui ou Non | Inconnu |

<br>

- Un problème de type NP qui est aussi de type P est **résoluble en temps polynomial.**
- Un problème NP-difficile qui est également NP-Complet est **vérifiable en temps polynomial.**

# III.&nbsp;&nbsp;Représentation formelle
On cherche à déterminer les tournées optimales de livraison pour un ensemble de camions partant d’un dépôt, visitant un ensemble de clients, et revenant au dépôt.
L’objectif est de minimiser la durée totale de tournée ou la date de retour du dernier camion, en respectant certaines contraintes (fenêtres de temps, capacités, compatibilités, etc.).
<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Paramètres :</h3>

Cette partie présente l’ensemble des informations connues avant la résolution.
Elle regroupe les différentes données disponibles qui serviront à la modélisation et à la résolution du problème.
Ces éléments montrent donc les données d’entrée du modèle. Nous allons représenter les différentes données reconnues dans le tableau suivant:

| Nom | Signification | Exemple |
|:----|:---------------|:---------|
| **V** | Liste des villes ou clients à livrer (le dépôt est la ville de départ) | V = {0, 1, 2, 3} |
| **K** | Nombre de camions disponibles | K = {1, 2} |
| **cᵢⱼ** | Temps entre la ville *i* et la ville *j* | c₁₂ = 12 min |
| **dᵢⱼ** | Distance entre la ville *i* et la ville *j* | d₁₅ = 5km |
| **Qᵢ** | Quantité à livrer au client *i* | Q₂ = 50 kg |
| **Cmₖ** | Capacité maximale du camion *k* | Cm₁ = 200 kg |
| **[aᵢ, bᵢ]** | Heure d’ouverture et de fermeture pour livrer le client *i* | [8h, 12h] |
| **tᵢ** | Temps nécessaire pour effectuer la livraison chez le client *i* | t₃ = 10 min |
| **type(i)** | Type de marchandise à livrer | type(2) = “fragile” |
| **comp(k)** | Type de marchandises que le camion *k* peut transporter | comp(1) = {“normal”, “fragile”} |
  
<br>

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> Variables de décision :</h3>

Les variables de décision permettent de marquer les choix à effectuer pour résoudre le problème.
Elles décrivent le comportement des camions, les trajets empruntés et les heures de livraison. Ces varaibles sont nécessaires pour construire une tournée optimale.

| Nom | Signification | Type |
|:----|:---------------|:---------------|
| **xᵢⱼᵏ** | 1 si le camion *k* passe de la ville *i* à la ville *j*, 0 sinon | Binaire |
| **tᵢ** | Heure d’arrivée du camion au point *i* | Réel positif |
| **qᵢᵏ** | Charge restante du camion *k* après avoir visité le point *i* | Réel positif |
| **yᵢᵏ** *(optionnel)* | 1 si le client *i* est servi par le camion *k*, 0 sinon | Binaire |

<br>

<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> Objectif et contraintes :</h3>

Notre modèle vise à **identifier des tournées de livraison minimisant le coût total**, tout en satisfaisant les différentes **contraintes**.
Il nous faut donc déterminer un **cycle hamiltonien minimal**, c’est-à-dire un circuit passant par toutes les villes une seule fois avant de revenir à la ville de départ.  

<div style="text-align:center;">
  
$$
\min \sum_{(i,j) \in C} c_{ij}
$$

</div>

Dans le cas où plusieurs camions sont disponibles, il est également possible de chercher à **minimiser la date de retour du dernier camion**, afin d’équilibrer au mieux la charge de travail entre les véhicules.  

Les **contraintes du modèle** assurent la cohérence et la possibilité réelle de réalisation des tournées.  
On s’assure ainsi que les livraisons suivent la logique du réseau ainsi que les **capacités physiques et temporelles**.  

Voici les principales contraintes retenues :  

- **Contrainte de visite unique :** chaque client doit être rendu **une seule fois** et par un **seul camion**.  
- **Contrainte de départ et de retour :** chaque camion **démarre du dépôt** au début du trajet et **retourne au dépôt à la fin du trajet**.  
- **Contrainte de capacité :** le volume total transporté par un camion **ne dépasse jamais la capacité maximale** de ce dernier.  
- **Contrainte de fenêtre de temps :** chaque livraison doit être réalisée **dans la fenêtre de temps permise**.  
- **Contrainte de cohérence temporelle :** le temps d’arrivée au client suivant dépend du **temps de trajet entre les clients**, ainsi que du **temps de service du premier client au second**.  
- **Contrainte de compatibilité :** certains camions ne peuvent être utilisés que pour livrer **certains types de marchandises**.  

- **Contrainte de visite unique :** chaque client doit être visité **une seule fois** et par un **seul camion**.
$$
\sum_{(j=1) \in n} x_{ij}^k=1\forall i = 1,...,n
$$
$$
\sum_{(i=1) \in n} x_{ij}^k=1\forall j = 1,...,n
$$

- **Contrainte de départ et de retour :** chaque camion **démarre du dépôt** au début de sa tournée et **revient au dépôt** à la fin du trajet.
<br>
$$
\sum_{(j)} x_{0j}^k=1
$$
$$
\sum_{(i)} x_{i0}^k=1
$$
- **Contrainte de capacité :** le volume total ou le poids des marchandises transportées par un camion **ne dépasse jamais la capacité maximale** autorisée.
$$
\sum_{(i,j)} Cm_{i} * x_{ij}^k \leq Q_i
$$

- **Contrainte de cohérence temporelle :** le **temps d’arrivée** à un client dépend du **temps de trajet** et du **temps de service** du client précédent.  


### Ajout de 2 contraintes supplémentaires

- **Contrainte de point de collecte :** chaque objet possède un **point de collecte spécifique** qui doit être visité avant sa livraison.  

Chaque objet **o** doit être ramassé à un point précis **p<sub>o</sub>** et livré à un point précis **l<sub>o</sub>**.

On suppose que le camion qui transporte l’objet doit passer par le point de collecte avant la livraison. 
Ici la variable de décision **x<sup>k</sup><sub>poj</sub>** doit etre égale à **y<sup>k</sup><sub>o</sub>**
avec **y<sup>k</sup><sub>o</sub>** = 1 ou 0 si le camion est affecté a ce cas spécifique
$$
\sum_{j} x_{p_oj}^k = y_o^k
$$
$$
\sum_{i} x_{p_oi}^k = y_o^k
$$

Chaque objet ne peut être livré que dans son point de livraison autorisé.
Autrement dit, on interdit toute livraison ailleurs.
$$
\sum_{i \neq l_o} x_{i l_o}^k = y_o^k
$$

- **Contrainte de fenêtre de temps :** chaque livraison doit être effectuée **dans la fenêtre horaire autorisée**.
$$
\ a_i \leq t_i \leq b_i
$$
  - Il est interdit de livrer en dehors de cette fenêtre.
  - Si le camion arrive en avance, il peut attendre sur place jusqu’à l’ouverture de la fenêtre temporelle.

# IV.&nbsp;&nbsp;Justification de la complexité
Pour finir, nous allons évaluer la complexité théorique.
Cette analyse permet de comprendre s’il est possible de résoudre le problème efficacement à l’aide d’un algorithme optimal, ou si celui-ci appartient à une classe de problèmes plus difficiles.
On cherche ainsi à savoir si le problème peut être résolu en temps polynomial, ou s’il est NP-difficile.

<h3 style="color:#1f4e79; font-style:italic; margin-top:0">Mise en relation :</h3>

Le problème de gestion de tournées de livraison étudié ici peut être rapproché du **TSP (Travelling Salesman Problem)**.  
Dans le TSP, un voyageur doit visiter une liste de villes **exactement une fois**, puis **revenir à son point de départ**, tout en **minimisant la distance totale parcourue**.  

Le TSP est un **problème combinatoire** :  
le nombre de solutions possibles (c’est-à-dire toutes les tournées envisageables) croît de manière **factorielle** avec le nombre de villes.  

$$
\text{Nombre de circuits possibles} = \frac{(n - 1)!}{2}
$$

Ainsi, plus le nombre de villes \(n\) augmente, plus le nombre de tournées possibles devient immense, rendant le **calcul exhaustif rapidement inabordable**.  

Cependant, du point de vue **algorithmique**, la complexité temporelle d’une méthode exacte (comme l’algorithme de Held–Karp) est **exponentielle**, généralement notée :

$$
O(n^2 \cdot 2^n)
$$

Ce qui signifie que **le temps de calcul double pratiquement à chaque ajout d’une nouvelle ville**, rendant la résolution exacte **impossible pour de grandes instances**.  

Le TSP (version d’optimisation) est donc classé comme un **problème NP-difficile**,  
c’est-à-dire qu’**aucun algorithme connu** ne permet de le résoudre efficacement en temps polynomial.  
Il n’est donc pas possible de vérifier ou de trouver rapidement une solution optimale sans tester un grand nombre de combinaisons.  

Notre problème de tournées de livraison, aussi appelé **VRP (Vehicle Routing Problem)**, est une **généralisation du TSP**, car il ajoute plusieurs contraintes réelles :  
- Plusieurs **camions** au lieu d’un seul voyageur,  
- Des **capacités limitées** pour chaque véhicule,  

Ainsi, le TSP est inclus dans le VRP, ce qui signifie que **résoudre le VRP implique savoir résoudre le TSP**.  
Puisque le TSP est **NP-difficile**, il en va de même pour le **VRP et toutes ses variantes**.  


<h3 style="color:#1f4e79; font-style:italic; margin-top:0"> Conséquences :</h3>

Le fait que notre problème appartienne à la classe NP-difficile a plusieurs conséquences importantes pour la résolution pratique.  

D’abord, cela signifie qu’il **n’existe pas de méthode exacte connue** capable de résoudre efficacement ce type de problème lorsque le nombre de villes ou de camions augmente.  
Les algorithmes exacts deviennent rapidement inexploitables, car leur temps de calcul croît **de manière exponentielle** avec la taille de l’instance.

Ainsi, pour des cas réels comportant des dizaines ou centaines de points de livraison, il devient nécessaire d’utiliser des **méthodes méta-heuristiques**.  
Ces méthodes ne garantissent pas d’obtenir **la solution optimale**, mais elles permettent d’obtenir **une solution de bonne qualité** en un **temps de calcul raisonnable**.
